### DataFrame 

2차원 테이블 데이터 구조, 엑실(스프레드 시트)와 비슷하다.

```python
import pandas as pd
import numpy as np
```

```python
sales_data = {
    '연도': [2015,2016,2017,2018,2019,2020],
    '판매량':[103,70,130,160,190,230],
    '매출':[500000,300000,400000,550000,700000,680000],
    '순이익':[370000,190000,300000,480000,600000,590000]
}
```

```python
sales_data = pd.DataFrame(sales_data)
```

```python
sales_data
```

```python
sales_dataframe = pd.DataFrame(sales_data, index = sales_data['연도'])
```

```python
sales_dataframe
```

```python
#columns -원하는 칼럼, index - 인덱스 칼럼 
```

```python
sales_dataframe = pd.DataFrame(sales_data, columns = ['판매량'],index = sales_data['연도'])
```

```python
sales_dataframe
```

```python
sales_dataframe = pd.DataFrame(sales_data, columns = ['판매량','매출','순이익'],index = sales_data['연도'])
```

```python
sales_dataframe
```

```python
#컬럼명 기준으로 데이터 출력
```

```python
sales_dataframe['판매량']
```

```python
# 행기준으로 데이터 조회
```

```python
sales_dataframe.iloc[1]
```

## 데이터(Column)추가

```python
sales_dataframe
```

```python
sales_dataframe['순이익율'] = (sales_dataframe['순이익'] / sales_dataframe['매출'] * 100)
```

```python
sales_dataframe['순이익율']
```

```python
sales_dataframe
```

```python
def check(n):
    if n > 80 :
        return '높음'
    else  :
        return '낮음'
```

```python
sales_dataframe['순이익율'].apply(check)
```

```python
sales_dataframe['순이익율_비교'] = sales_dataframe['순이익율'].apply(check)
```

```python
sales_dataframe
```

```python
#인덱스에 조건가능
```

```python
sales_dataframe[sales_dataframe['매출'] > 300000]
```

```python
#해당 조건에 일치한는 것만 데이터 변경
```

```python
sales_dataframe['테스트1'] = np.where(sales_dataframe['판매량']> 200, 0, sales_dataframe['판매량'])
```

```python
sales_dataframe
```

```python
#해당 조건에 일치하는 것만 바꾸고 나머지는 NaN
```

```python
sales_dataframe['테스트2'] = sales_dataframe[sales_dataframe['판매량']<100]['판매량']+50
```

```python
sales_dataframe
```

```python
#열 추가 
```

```python
sales_dataframe.loc[2021] = [100, 720000, 650000, 30, '낮음', 0, 0]
```

```python
sales_dataframe
```

```python
sales_dataframe.loc[2022] = sales_dataframe.loc[2021]
```

```python
sales_dataframe
```

### GroupBy

```python
df = pd.DataFrame(np.random.randint(1,100,size=(8,2)),
                 index = [['A창고','A창고','A창고','A창고','B창고','B창고','B창고','B창고'],
                          ['사과','배','바나나','사과','사과','배','바나나','배']],
                 columns = ['판매','재고'])
```

```python
df
```

```python
df.index.names=['창고명','상품명']
```

```python
df
```

```python
df.groupby('창고명').sum()
```

```python
df.groupby('상품명').sum()
```

```python
df.groupby(['창고명','상품명']).sum()
```

### sort_values

```python
df.sort_values(by='판매',ascending =True)
```

```python
df.sort_values(by='판매',ascending =False)
```

```python
df.sort_values(by=['판매','재고'],ascending =[False,)
```
